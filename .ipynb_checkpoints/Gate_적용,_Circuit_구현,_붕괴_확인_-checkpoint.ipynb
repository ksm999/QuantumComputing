```python
from qiskit import QuantumCircuit, execute, BasicAer
from qiskit.visualization import plot_histogram
```

```python
qc = QuantumCircuit(2,2)
qc.h(0)
qc.ry(pi/4,0)
#qc.cx(0,1)
qc.draw('mpl')
state = Statevector.from_instruction(qc)
plot_bloch_multivector(state)
```

```python
qc = QuantumCircuit(2,2)
qc.h(0)
qc.ry(pi/4,0)
qc.cx(0,1)
qc.measure([0,1],[0,1])
qc.draw('mpl')
```

```python
backend = BasicAer.get_backend('qasm_simulator')
job = execute(qc, backend, shots=1024)
result = job.result()
counts = result.get_counts()
counts
```

```python
plot_histogram(counts)
```

```python
qc = QuantumCircuit(3,3)
qc.h(range(3))
qc.cx(0,1)
qc.cx(0,2)
qc.rz(pi/4,1)
qc.y(2)
qc.draw('mpl')
```

```python
state = Statevector.from_instruction(qc)
plot_bloch_multivector(state)
```

```python
qc.measure_all()
qc.draw('mpl')
```

```python
backend = BasicAer.get_backend('qasm_simulator')
job = execute(qc, backend, shots=1024)
result = job.result()
counts = result.get_counts()
counts
```

```python
plot_histogram(counts)
```

```python
qc= QuantumCircuit(4,2)
qc.x(0)
qc.x(1)
qc.barrier()
qc.cx(0,1)
qc.cx(1,2)

qc.ccx(0,1,3)
qc.barrier()
qc.measure(2,0)
qc.measure(3,1)
qc.draw('mpl')
```

```python
backend = BasicAer.get_backend('qasm_simulator')
job = execute(qc, backend, shots=1024)
result = job.result()
counts = result.get_counts()
print(counts)
plot_histogram(counts)
```
