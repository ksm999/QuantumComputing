## Quantum Computing's entire workflow

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

# Use Aer's AerSimulator
simulator = AerSimulator()

# Create a Quantum Circuit acting on the q register
# QuantumCircuit(qubit, bit)
circuit = QuantumCircuit(2, 2)

circuit.h(0) # Add a H gate on qubit 0

circuit.cx(0,1) # Add a CX gate on control qubit 0 and target qubit 1

# Map the quanutm measurement to the classical bits
circuit.measure([0, 1],[0, 1])

# Compile
#circuit.draw('mpl')

compiled_circuit = transpile(circuit, simulator)

job = simulator.run(compiled_circuit, shots=1000)

result = job.result()

counts = result.get_counts(compiled_circuit)
print('\nTotal count for 00 and 11 are : ', counts)
plot_histogram(counts)
```

코드를 보면 양자회로(QuantumCircuit)를 정의한 다음, 회로에 h 와 cx를 취한 다음, 측정(measure)을 한 듯 하다.
또한 처음에 정의했던 simulator와 회로(circuit)를 변환(transpile) 하여 결과를 보인 것 같다.
